```python
import numpy as np
x=np.arange(1,3)
y=x**2
plt.plot(x,y,c="red",marker="o",linestyle="-")
plt.show()
```

```python
plt.plot(x,y,c="black",marker="+",linestyle="-.",label="x and x* 2")
plt.plot(x,y*2,c="y",marker="d",linestyle="-",label="y and y*2")
plt.legend(loc="center")
plt.show()
```

```python
x=np.linspace(0,10,40)
y=np.sin(x)
plt.plot(x,np.cos(x),label="cos(x)",marker="*", c="m")
plt.plot(x,y,label="sos(x)",marker="^", c="c")
plt.legend()
plt.show()
```

```python
fig,area=plt.subplots(2,2)
x=np.linspace(1,10,20)
area[0][0].plot(x,x)
area[0][0].set_title("Line")
area[0][1].plot(x,x**2)
area[0][1].set_title("square")
area[1][0].plot(x,x**3)
area[1][0].set_title("cube")
area[1][1].plot(x,np.log(x))
area[1][1].set_title("log")
plt.show()
```

```python
gp1=plt.subplot2grid((3,3),(0,0),colspan=2)
gp2=plt.subplot2grid((3,3),(0,2),rowspan=2)
gp3=plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=1)
x=np.arange(1,20)
gp1.plot(x,np.log(x))
gp2.plot(x,np.exp(x))
gp3.plot(x,np.cos(x))
plt.show()
```

# bar graph

```python
subjects=["C","Python","C++","Java","R"]
marks=[90,67,78,65,89]

plt.bar(subjects,marks,color=["c","m","b","y","r"],width=0.3)
plt.xlabel("SubjectsNAmes")
plt.ylabel("Marks outof 100")
plt.show()
```

```python
subjects=["C","Python","C++","Java","R"]
marks=[90,67,78,65,89]
plt.barh(subjects,marks,color=["c","m","b","y","r"])
plt.xlabel("SubjectsNAmes")
plt.ylabel("Marks outof 100")
plt.text(40,"R","Low")
plt.text(98,"Python","High")
plt.show()
```

```python
subjects=["C","Python","C++","Java","R"]
marks=[90,67,78,65,89]
plt.barh(subjects,marks,color=["c","m","b","y","r"])
plt.bar(subjects,marks)
plt.xlabel("SubjectsNAmes")
plt.ylabel("Marks outof 100")

plt.show()
```

# Pie charts

```python
plt.pie(marks,labels=subjects,colors=["r","c","m","y","b"],radius=1.5,autopct='%1.2f%%')
plt.show()
                                 
```
